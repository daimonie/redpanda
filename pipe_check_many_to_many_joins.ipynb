```python
import pandas as pd
import numpy as np
```

```python
df_left = pd.DataFrame.from_dict({
    'category': np.random.choice(['apples', 'oranges', 'avocados'], size=100),
    'id': [_ for _ in range(100)],
    'kpi': np.random.randint(0,high=100, size=100)
})
df_left.head(10)
```

```python

df_right = pd.DataFrame.from_dict({
    'category': np.random.choice(['apples', 'oranges', 'avocados'], size=100),
    'id': [_ for _ in range(100)],
    'kpi2': np.random.randint(0,high=100, size=100)
})
df_right.head(10)
```

```python
def _assert_type(comparison, left, right):
    if comparison == "eq":
        assert left == right
    elif comparison == "leq":
        assert left <= right
    elif comparison == "le":
        assert left < right
    elif comparison == "geq":
        assert left >= right
    elif comparison == "ge":
        assert left > right

def check_many_to_many(df, n=1, m=1, comparison='eq'):
    if not hasattr(check_many_to_many, "counter"):
        setattr(check_many_to_many, "counter", 0)
    
    counter = getattr(check_many_to_many, "counter")

    if counter%2 == 0:
        # We are before a merge
        setattr(check_many_to_many, "size_before", len(df))
    else:
        # We are after a merge
        setattr(check_many_to_many, "size_after", len(df))
        
        ratio = getattr(check_many_to_many, "size_before") / len(df)

        print(f"Comparing ratio {ratio} against ratio for {n}:{m} join.")
        _assert_type(comparison, ratio, n/m)

    setattr(check_many_to_many, "counter", counter+1)

    return df
```

```python
# join that is 1:1
(
    df_left
    .pipe(check_many_to_many)
    .merge(df_right, how='left')
    .pipe(check_many_to_many)
)
```

```python
# join that should fail
(
    df_left
    .pipe(check_many_to_many)
    .merge(df_right, how='left', on="category")
    .pipe(check_many_to_many)
)
```
